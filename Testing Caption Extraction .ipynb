```python
#test using extracting zh captions youtube video

yt = YouTube('https://www.youtube.com/watch?v=5VjD8d_2ZLo')
yt.streams
yt.captions
caps = yt.captions
zh_caps = caps['zh']
xml_caps = zh_caps.xml_captions
xml_caps
```

```python
#converting xml
xml_caps = zh_caps.xml_captions

import xml.etree.ElementTree as ET

# Parse the XML captions using Element Tree as I couldnt get the srt function in pytube to work. 
xml = ET.fromstring(xml_caps)

# Extract the text content from the XML captions
captions_text = ""
for text in xml.iter('p'):
    caption = text.text
    if caption:
        captions_text += caption + " "

# Print the extracted captions text
print(captions_text)
```
