```python
#creating a smaller df to test function for translating whole columns
df2 = df.head(10).copy()
df2
```

```python
#applying translation to all titles
df2['translated titles'] = df2['titles'].apply(lambda x: (time.sleep(3), translator.translate(x, src='zh-TW', dest='en').text))

df2.head()
```

```python
#creating a translate function

def translate(text):
    time.sleep(3)
    translator = Translator()
    translation = translator.translate(text, src='zh-TW', dest='en')
    return translation.text
```

```python
#testing
text = '是您獲取商業思維、加密貨幣世界重點資訊的一站式頻道'
translate(text)
```

```python
#reload df2 and test on titles column only
df2['translated titles'] = df2['titles'].apply(translate)
```

```python
#testing on more columns 
columns_to_translate = ['tags', 'descriptions']

for column in columns_to_translate:
    df2[f'translated {column}'] = df2[column].apply(translate)
```
