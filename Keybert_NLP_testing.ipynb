```python
from keybert import KeyBERT

doc = """
         Supervised learning is the machine learning task of learning a function that
         maps an input to an output based on example input-output pairs. It infers a
         function from labeled training data consisting of a set of training examples.
         In supervised learning, each example is a pair consisting of an input object
         (typically a vector) and a desired output value (also called the supervisory signal).
         A supervised learning algorithm analyzes the training data and produces an inferred function,
         which can be used for mapping new examples. An optimal scenario will allow for the
         algorithm to correctly determine the class labels for unseen instances. This requires
         the learning algorithm to generalize from the training data to unseen situations in a
         'reasonable' way (see inductive bias).
      """

#Extracting the keywords
kw_model = KeyBERT()
keywords = kw_model.extract_keywords(doc)
```

```python
#Set the keywords length as ngram(1,1)
kw_model.extract_keywords(doc, keyphrase_ngram_range=(1, 1), stop_words=None)
```

```python
#Set the keywords length as ngram(1,2)
kw_model.extract_keywords(doc, keyphrase_ngram_range=(1, 2), stop_words=None)
```

```python
# Use Maximum Marginal Relevance to extract high diversity keywords
kw_model.extract_keywords(doc, keyphrase_ngram_range=(3, 3), stop_words='english',
                              use_mmr=True, diversity=0.7)
```

```python
# Use Maximum Marginal Relevance to extract low diversity keywords
kw_model.extract_keywords(doc, keyphrase_ngram_range=(3, 3), stop_words='english',
                              use_mmr=True, diversity=0.2)
```

```python
# Sentence Transformer Model
from sentence_transformers import SentenceTransformer

sentence_model = SentenceTransformer("paraphrase-MiniLM-L6-v2")
kw_model = KeyBERT(model=sentence_model)
```

```python
# Set the keywords length as ngram(1,1)
kw_model.extract_keywords(doc, keyphrase_ngram_range=(1, 1), stop_words=None)
```

```python
# Set the keywords length as ngram(1,2)
kw_model.extract_keywords(doc, keyphrase_ngram_range=(1, 2), stop_words=None)
```

```python
# Use Maximum Marginal Relevance to extract high diversity keywords
kw_model.extract_keywords(doc, keyphrase_ngram_range=(3, 3), stop_words='english',
                              use_mmr=True, diversity=0.7)
```

```python
# Use Maximum Marginal Relevance to extract low diversity keywords
kw_model.extract_keywords(doc, keyphrase_ngram_range=(3, 3), stop_words='english',
                              use_mmr=True, diversity=0.2)
```
